```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('parkinsons_updrs.data')
```

```python
df
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.describe()
```

```python
decimal_parts = (df['test_time'] % 1 * 10000).round().astype(int)

# Find index of the largest decimal part
max_idx = decimal_parts.idxmax()
max_decimal = decimal_parts[max_idx]
original_value = df.loc[max_idx, 'test_time']

print(f"Largest decimal part (as int): {max_decimal}")
print(f"At index: {max_idx}")
print(f"Original value: {original_value}")
```

```python
sns.set(style="whitegrid")
sns.pairplot(df, diag_kind="kde", markers='o', hue='subject#', palette='husl')
plt.savefig('pairplot.png')
```
